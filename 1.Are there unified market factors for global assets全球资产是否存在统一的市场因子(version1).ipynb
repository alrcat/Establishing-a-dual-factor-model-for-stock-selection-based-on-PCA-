### 1.股票指数

#### 数据处理

```python
#矩阵化和归0化
stocks = list(d1data.keys())

dates = list(d1data[stocks[0]].index)

data_array = np.zeros((len(dates), len(stocks)))

for i, stock in enumerate(stocks):
    data_array[:, i] = d1data[stock]['open'].values

df = pd.DataFrame(data_array, index=[str(date) for date in dates], columns=stocks)

df.fillna(0, inplace=True)

# DataFrame
pivot_df = df
#print(pivot_df)

#z-score by line
def standardize_row(row):
    row_mean = row.mean()
    row_std = row.std()
    return (row - row_mean) / row_std

pivot_df_standardized = pivot_df.apply(standardize_row, axis=1)

print(pivot_df_standardized)

pivot_df = pivot_df_standardized
```

#### pca模型

```python
data_array = pivot_df.to_numpy()

pca = PCA(n_components=3)

pca.fit(data_array)

transformed_data = pca.transform(data_array)

#DataFrame
transformed_df = pd.DataFrame(transformed_data, index=pivot_df.index, columns=['PCA1', 'PCA2', 'PCA3'])

print(transformed_df)
print(type(transformed_df))
```

```python
#calculate
explained_variance = pca.explained_variance_

total_variance = np.sum(explained_variance)

explained_variance_ratio = explained_variance / total_variance

cumulative_explained_variance_ratio = np.cumsum(explained_variance_ratio)

#print
for i, ratio in enumerate(explained_variance_ratio):
    print(f'PCA{i+1} : {ratio:.2f}')

#sum
print(f'SUM：{cumulative_explained_variance_ratio}')
```

```python
weights = pca.components_

for i, pc_weights in enumerate(weights):
    print(f'PCA{i+1}：')
    for j, weight in enumerate(pc_weights):
        print(f'component{j+1} ：{weight:.4f}')
```
